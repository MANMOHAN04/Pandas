```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("imdb-top-1000.csv")
```

```python
movies.head(3)
```

```python
movies['Genre'].unique()
```

```python
geners = movies.groupby('Genre')
```

```python
#Applying some aggregates function on groupby objects
geners.sum()
```

```python
#find the top 3 genres by total earnings
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=)
```
